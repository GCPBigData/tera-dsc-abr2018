#Minha primeira rede neural em Keras

## Pacotes necessários

```

from keras.models import Sequential
from keras.layers import Dense
import numpy as np 
# Semente fixa para tornar o código reprodutível
numpy.random.seed(42)

import matplotlib.pyplot as plt
import pandas as pd
```

```
# Histórico / Prova
input = np.array([[0.7, 0.6], [0.9, 0.2], 
                  [0.2, 0.7], [0.9, 0.9], 
                  [0.7, 0.7], [0.8, 0.9],
                  [1.0, 0.7], [0.5, 0.9],
                  [0.2, 0.7], [0.9, 0.7],
                  [0.9, 0.8], [0.4, 0.9],
                  [0.2, 0.4], [0.4, 0.1],
                  [0.1, 0.2], [0.1, 0.3]])

output = np.array([1, 0, 
                   0, 1, 
                   1, 1, 
                   1, 1,
                   0, 1, 
                   1, 0,
                   0, 0,
                   0, 0])

df = pd.DataFrame(data=input, columns=['x', 'y'])

df.plot.scatter(x='x', y='y', xlim =[0,1.0], ylim=[0,1.0])
```

## Criando o modelo

```
model = Sequential()
model.add(Dense(3, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

## Compilando o modelo em Keras

```
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## Fit do modelo

```
# Fit the model
X = df.values
Y = output
model.fit(X, Y, epochs=10, batch_size=10)
```

## Avaliando o modelo

```
# evaluate the model
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
